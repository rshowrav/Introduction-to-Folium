Import the folium library

```python
import folium
```

Using the code below you can quickly get a world map. Note the capablity to zoom out and in. 

```python
world_map = folium.Map()
world_map
```

Now lets say that we want to research a given location. For us we will use Dhaka, Bangladesh.

```python
bangladesh = folium.Map(location=[23.777176, 90.399452], control_scale=True)
bangladesh
```

Note the above map is in the native Bangladesh language. We will need to pick a tile using the below to handle this.

```python
bangladesh = folium.Map(location=[23.777176, 90.399452], control_scale=True, tiles='stamenterrain')
bangladesh
```

Now we are getting a bit better. But lets zoom in. 

```python
bangladesh = folium.Map(location=[23.777176, 90.399452], zoom_start = 14, control_scale=True, tiles='stamenterrain')
bangladesh
```

Now lets add a marker to a point on the map. Lets put it at the location on the map titled Chandrima Uddan.

```python
folium.Marker([23.7667,90.3781]).add_to(bangladesh)
bangladesh
```

Notice the marker covered up the name of our location. We need to add a popup for user to see the name when they click the marker.

```python
folium.Marker([23.7667,90.3781], popup = "Chandrima Uddan").add_to(bangladesh)
bangladesh
```

Lastly lets make that marker more noticable.

```python
folium.Marker([23.7667,90.3781], popup = "Chandrima Uddan", icon=folium.Icon(color="blue", icon="info-sign")).add_to(bangladesh)
bangladesh
```
